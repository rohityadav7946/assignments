```python
#QUESTION 1
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
```

```python
df.head()
```

```python
sns.countplot(x='Date', data=df)
plt.title("1.Count plot of Date")
```

```python
sns.distplot(df['MinTemp'])
plt.title("2.Dist plot of MinTemp")
```

```python
sns.countplot(x='MaxTemp', data=df)
plt.title("3.Count plot of MaxTemp")
```

```python
sns.distplot(df['Rainfall'])
plt.title("4.Dist plot of Rainfall")
```

```python
sns.boxplot(x= 'WindSpeed9am', y='WindSpeed3pm',data=df, hue='RainToday')
plt.title("5.plot of  'WindSpeed9am','WindSpeed3pm'")
```

```python
sns.jointplot(x= 'Humidity9am', y='Humidity3pm',data=df)
plt.title("6.plot of  'Humidity9am','Humidity3pm'")
```

```python
sns.boxplot(x='Evaporation', y='Sunshine',data=df)
plt.title("7.Box plot of 'Evaporation','Sunshine',")
```

```python
sns.jointplot(x='Evaporation', y='Sunshine',data=df, alpha=0.2)
plt.title("8. plot of 'WindGustDir','WindGustSpeed'")
```

```python
df['RISK_MM'].plot(kind='barh')
plt.title("9.plot of RISK_MM")
```

```python
sns.violinplot(x= 'Pressure9am', y='RainTomorrow',data=df)
plt.title("10.plot of  'Pressure9am','RainTomorrow'")
```

```python
sns.violinplot(x= 'WindDir9am', y='WindSpeed9am',data=df, hue='RainTomorrow',inner='stick')
plt.title("11.plot of  'WindDir9am','WindSpeed9am'")
```

```python
sns.boxplot(x= 'WindDir9am', y='Temp3pm',data=df,palette='Set2')
plt.title("12.plot of  'WindDir9am','Temp3pm'")
```

```python
sns.violinplot(x= 'Temp9am', y='Temp3pm',data=df,hue='RainToday' ,split= True, inner='stick',palette='Set2')
plt.title("13. plot of  'Temp9am','Temp3pm'")
```

```python
sns.violinplot(x= 'RainToday', y='WindSpeed9am',data=df, hue='RainTomorrow',split=True,inner='stick')
plt.title("14. plot of  'RainToday','WindSpeed9am'")
```

```python
sns.jointplot(x= 'Pressure3pm', y='Humidity3pm',data=df)
plt.title("15.plot of  'Pressure3pm','Humidity3pm'")
```
