```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
df1 = pd.read_csv("D:\csv files/test.csv")
df2 = pd.read_csv("D:\csv files/train.csv")
```

```python
frames = [df2,df1]
```

```python
df = pd.concat(frames)
```

```python
df.head()
```

```python
df.shape
```

```python
df['Survived']=df['Survived'].dropna()
```

```python
df.head()
```

```python
Has_Cabin=~df['Cabin'].isnull()
df.loc[:,'Has_Cabin'] = Has_Cabin
```

```python
df.head()
```

```python
import re
```

```python
df["Title"]=df.Name.apply(lambda x : re.search('([A-Z][a-z]+)\.',x).group(1))
df["Title"].unique()
```

```python
list = ["Don","Rev","Dr","Mme","Ms","Major","Lady","Sir","Mlle","Col","Capt","Countess","Jonkheer"]
df["Title"] = df["Title"].replace(list,"Special")
```

```python
df["Title"].unique()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Embarked'] = df.Embarked.fillna('S')
```

```python
df['Fare'] = df.Fare.fillna(df.Fare.median())
```

```python
df['Age'] = df.Age.fillna(df.Age.median())
```

```python
df.head()
```

```python
df['Categorical_Age']=pd.qcut(df['Age'],q=4,labels=False)
df['Categorical_Fare']=pd.qcut(df['Fare'],q=4,labels=False)
df.head()
```

```python
del df['Name']
del df['Cabin']
del df['Age']
del df['PassengerId']
del df['SibSp']
del df['Ticket']
del df['Parch']
del df['Fare']
```

```python
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
labels=df['Survived'].values
```

```python
del df['Survived']
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
(x_train,x_test,y_train,y_test)=train_test_split(df,labels,test_size=0.3)
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LogisticRegression()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
result = pd.DataFrame({"Predicted":pred,"Actual":y_test})
```

```python
result
```

```python
model.score(x_test,y_test)
```
